```python
import pandas as pd
from sklearn.datasets import load_digits

# explore in pandas
dataset = load_digits()
pandas_df = pd.DataFrame(dataset.data,
                         columns=dataset.feature_names)
pandas_df['target'] = dataset.target

print(pandas_df.shape)
print(pandas_df.columns)
print(pandas_df.head())
print(pandas_df.values.min(), pandas_df.values.max())
print(pandas_df.iloc[:, -1].unique())
```

```python
# explore in spark
import pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("PySpark_Dataset_Inspection").getOrCreate()
spark_df = spark.createDataFrame(pandas_df)
spark_df.show(5)
print(spark_df.count(), len(spark_df.columns))
print(spark_df.printSchema())
spark_df.summary().show()
(spark_df.select(spark_df.columns[-1])
 .distinct()
 .orderBy(spark_df.columns[-1])
 .show()
)
```
